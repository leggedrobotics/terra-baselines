```python

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300

import sys
sys.path.append("..")
from utils.helpers import load_pkl_object

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
env_names = ['Terra']

fig, axs = plt.subplots(figsize=(20, 12), nrows=2, ncols=1)

for i, e in enumerate(env_names):
    ax = axs.flatten()[i]
    
    results = load_pkl_object(f"../agents/{e}/ppo.pkl")
    steps = results["log_steps"]
    cum_rew = results["log_return"]
    ax.plot(steps, cum_rew, label="PPO - lr: 5e-04")

    ax.set_xticks(steps)
    ax.xaxis.set_major_locator(plt.MaxNLocator(4))
    ax.yaxis.set_major_locator(plt.MaxNLocator(4))
    
    if i in [6, 7, 8]:
        ax.set_xlabel('Env Steps')
    if i in [0, 3, 6]:
        ax.set_ylabel('Return')
    ax.set_title(f'{e}')
    if i == 8:
        ax.legend(fontsize=16)
    print(e, cum_rew[-1])

fig.tight_layout()
```
